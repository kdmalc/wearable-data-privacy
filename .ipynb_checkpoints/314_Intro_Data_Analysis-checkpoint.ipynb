> Find entropy in order to characterize attributes by most informative

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import rel_entr
```

```python
################## DATA FRAME ##################
datafile = 'Data/Fitbit_Kaggle/dailyActivity_merged.csv'
fields = ['Id', 'ActivityDate', 'TotalSteps', 'TotalDistance', 'TrackerDistance', 
          'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance', 
          'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'Calories']

df = pd.read_csv(datafile, skipinitialspace=True, usecols=fields)
```

```python
df.head()
```

```python
all_IDs = df.Id.unique()
num_users = len(all_IDs)
```

```python
calorie_bins = [0] * num_users
for idx, user in enumerate(all_IDs):
    temp_df = df.loc[df['Id'] == user]
    edited_df = temp_df[temp_df['Calories'] !=0]
    
    caloric_series = edited_df["Calories"]
    count, division = np.histogram(caloric_series)
    calorie_bins[idx] = count/sum(count)
    #caloric_series.hist(bins=division)
```

```python
distance_matrix = np.zeros((num_users, num_users))

for i in range(int(num_users-1)):
    for j in range(int(num_users-1)):
        distance_matrix[i, j] = sum(rel_entr(np.array(calorie_bins[i]),np.array(calorie_bins[j])))
```

```python
# box =[P(green),P(blue),P(red),P(yellow)]
box_1 = [0.25, 0.33, 0.23, 0.19]
box_2 = [0.21, 0.21, 0.32, 0.26]
  
import numpy as np
from scipy.special import rel_entr
  
def kl_divergence(a, b):
    return sum(a[i] * np.log(a[i]/b[i]) for i in range(len(a)))
    
print('KL-divergence(box_1 || box_2): %.3f ' % kl_divergence(box_1,box_2))
print('KL-divergence(box_2 || box_1): %.3f ' % kl_divergence(box_2,box_1))
  
# D( p || p) =0
print('KL-divergence(box_1 || box_1): %.3f ' % kl_divergence(box_1,box_1))
  
print("Using Scipy rel_entr function")
box_1 = np.array(box_1)
box_2 = np.array(box_2)
  
print('KL-divergence(box_1 || box_2): %.3f ' % sum(rel_entr(box_1,box_2)))
print('KL-divergence(box_2 || box_1): %.3f ' % sum(rel_entr(box_2,box_1)))
print('KL-divergence(box_1 || box_1): %.3f ' % sum(rel_entr(box_1,box_1)))
```
