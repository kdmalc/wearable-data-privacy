```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Priv = pd.read_csv('Utility_SanData.csv')
Priv.head()
```

```python
sns.barplot(x='Epsilon', y='maxErr', hue='blockSize', data=Priv)
```

```python
sns.barplot(x='Epsilon', y='L1', hue='blockSize', data=Priv)
```

```python
sns.barplot(x='Epsilon', y='KL', hue='blockSize', data=Priv)
```

```python
#Average distance in minutes for the activites
ax = sns.barplot(x='Epsilon', y='AverageProfileDistance', hue='blockSize', data=Priv)
ax.set_yscale('log')
```

Some observations:
1. for large privacy regimes (e.g., \epsilon>5.0) the smaller block size we use the better usability we preserve. 
2. with this small dataset we need large privacy values to preserve usability (\epsilon >=10). For example, for ep=10 and blockSize=5, we loose 200 mins of activity on averge per individual (over 30 days of recording).
